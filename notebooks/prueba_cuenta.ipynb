```python
import pandas as pd
import numpy as np

# Cargar el archivo Excel
df = pd.read_excel('test.xlsx', engine='openpyxl', skiprows=4, header=None)
```

```python
df
```

```python
# Renombrar columnas para facilitar el manejo
df.columns = ["fecha_contable", "fecha", "concepto", "importe", "saldo"]

# Limpieza y conversión de datos
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce').dt.strftime('%Y-%m-%d')
df['importe'] = pd.to_numeric(df['importe'], errors='coerce')
df['saldo'] = pd.to_numeric(df['saldo'], errors='coerce')

final_df = df[['fecha', 'concepto', 'importe']].copy()
final_df.dropna(subset=['fecha', 'concepto', 'importe'], inplace=True)
final_df['origen'] = 'cuenta'
final_df['entidad'] = 'bankinter'

final_df
```
